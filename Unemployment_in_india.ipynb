**EXPLORATORY DATA ANALYSIS**

Barplot of region with estimated umemloyment rate to check the unemployement region wise

```python
data.columns
```

```python
# Sorting the DataFrame by the Estimated Unemployment Rate in descending order
data_sorted = data.sort_values(by=" Estimated Unemployment Rate (%)", ascending=False)
data_sorted.head()
```

```python
#plotting the barchart
plt.figure(figsize=(10,5))
plt.bar(data_sorted['Region'],data_sorted[' Estimated Unemployment Rate (%)'],color='red')
plt.title('Average unemployement Rate')
plt.xlabel('Region')
plt.ylabel('Estimated unemployement Rate %')
# Rotating the x-axis labels
plt.xticks(rotation=45, ha='right')
plt.tight_layout()# Adjust layout to make room for rotated labels
#displaying the chart
plt.show()
```



*   Region with highest umemployement rate is puduchery
*   Region with lowest unemployement rate is Assam

BOXPLOT OF ESTIMATED UMEMPLOYEMENT RATE BY REGION

```python
plt.figure(figsize=(20,15))
sns.boxplot(x='Region',y=' Estimated Unemployment Rate (%)',data=data)
plt.title('Estimtated unemployment rate by region')
plt.xlabel('Region')
plt.ylabel('Estimated unemployement Rate')
plt.xticks(rotation=45,ha='right')#ha =horizontal alighnment.
plt.tight_layout()
plt.show()
```

PIECHART OF AREA WISE UNEMPLOYEMENT RATE

```python
plt.figure(figsize=(10,5))
plt.pie(data['Area'].value_counts(),labels=data['Area'].value_counts().index,autopct='%1.1f%%')
plt.title('Area wise unemployement rate')
plt.show()
```

Urban area has highest unemployment rate

Heat map of correlation matrix

```python
plt.figure(figsize=(10,5))
corr = data[[' Estimated Unemployment Rate (%)', ' Estimated Employed', ' Estimated Labour Participation Rate (%)']].corr()
sns.heatmap(corr,annot=True,linewidth=0.5,cmap='coolwarm')#CMAP MEANS COLORMAP
plt.show()
```

As unemployment rates increase, the number of employed individuals tends to decrease, indicating an inverse relationship.
Variations in unemployment rates have little impact on overall labor force participation.
A rise in employment may slightly boost labor force participation, but this effect is not very pronounced.

## ***summary of findings***

```python
avg_unemployement_over_time = data.groupby(' Date')[' Estimated Unemployment Rate (%)'].mean().sort_values(ascending=False).reset_index()
avg_unemployement_over_region = data.groupby('Region')[' Estimated Unemployment Rate (%)'].mean().sort_values(ascending=False).reset_index()
avg_people_employed_over_region = data.groupby('Region')[' Estimated Employed'].mean().sort_values(ascending=False).reset_index()
avg_labour_part_rate_over_region = data.groupby('Region')[' Estimated Labour Participation Rate (%)'].mean().sort_values(ascending=False).reset_index()
avg_unemployement_area_wise = data.groupby('Area')[' Estimated Unemployment Rate (%)'].mean().sort_values(ascending=False).reset_index()
highest_unemployement_rate_month = data.groupby(' Date')[' Estimated Unemployment Rate (%)'].max().sort_values(ascending=False).reset_index()

# print all the dataframes
print("Average Unemployment Rate over time\n", avg_unemployement_over_time)
print("\nAverage Unemployment Rate over Region\n", avg_unemployement_over_region)
print("\nAverage People employed over Region\n", avg_people_employed_over_region)
print("\nAverage Labour Participation Rate over Region\n", avg_labour_part_rate_over_region)
print("\nAverage Unemployment Rate over Area\n", avg_unemployement_area_wise)
print("\nHighest Unemployment Rate Month wise\n", highest_unemployement_rate_month)
```
